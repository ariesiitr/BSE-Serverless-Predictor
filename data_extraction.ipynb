```python
close=pd.DataFrame()
columns_list=[];
for row,company in quandl_codes.iterrows():
    data=quandl.get('BSE/'+company['code'],start_date="2010-01-01")
    company_name=company['name']
    close[company_name]=data['Close']
    columns_list.append(company_name)
    data.to_excel('data/'+company_name+'.xlsx')
    close.to_excel('data/close.xlsx')
    #if(company['to_date']==date.today()):
    #    close
```
